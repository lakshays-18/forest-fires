```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

## fix column name spaces

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']] = df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
```

## EDA and Feature Engineering

```python
df_copy = df.drop(columns=['day','month','year'],axis=1)
```

```python
df_copy
```

## Converting obejcts to numeric for classes column

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
plt.style.use('seaborn')
df_copy.hist(bins=50,figsize=(20,16))
plt.show()
```

## plotting pie chart for classes

```python
percent = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(10,10))
plt.pie(percent,labels=['fire','not fire'],autopct='%1.1f%%')
plt.title('Pie Chart for classes')
plt.show()
```

## Monthly fire analysis

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
plt.figure(figsize=(10,10))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.xlabel('Months',weight='bold')
plt.ylabel('Number of fires',weight='bold')
plt.show()
```
